# The function probability_general is the most general formula. It returns probability of muon oscillations with/without matter effects, for NO/IO

$ \textbf{Example #1: Comparison of exact probility in matter, in vacuum and leading approximation}$

```python
energy = np.linspace(0.2, 2, 1000)
U = PMNS()

prob_mue_NO = probability_vacuum(U=U, E=energy, beta=0)
prob_mue_NO_matter = probability_general(E=energy, beta=0)
prob_mue_approx = probability_vacuum_approx(beta=0, E=energy)
```

```python
plt.plot(energy, prob_mue_NO_matter, label='Exact formula in matter')
plt.plot(energy, prob_mue_NO, label='Exact formula in vacuum')
plt.plot(energy, prob_mue_approx, label='Leading approximation')

plt.ylabel(r'$P (\nu_\mu \rightarrow \nu_e)$')


plt.legend()
```

$ \textbf{Example #3}:$ Comparison of NO and IO in vacuum and in matter

```python

U = PMNS(delta_cp=-90)
energy = np.linspace(0.25, 2, 1000)

prob_mue_NO = probability_vacuum(U, beta=0, E=energy)
prob_mue_IO = probability_vacuum(U, beta=0, E=energy, dm2_32=-dm2_32_BF)
plt.plot(energy, prob_mue_NO, label='NO, vacuum')
plt.plot(energy, prob_mue_IO, label='IO, vacuum')

plt.ylabel(r'$P (\nu_\mu \rightarrow \nu_e)$')
plt.xlabel('energy, GeV')
plt.legend()
plt.savefig("Vacuum")
```

```python
prob_mue_matter_NO = probability_general(E=energy, beta=0, delta_cp=radians(90))
prob_mue_matter_NO2 = probability_general(E=energy, beta=0, delta_cp=radians(-90))
prob_mue_matter_IO = probability_general(E=energy, beta=0, delta_cp=radians(90), MO='IO')
prob_mue_matter_IO2 = probability_general(E=energy, beta=0, delta_cp=radians(-90), MO='IO')

prob_mue_matter_anti_NO = probability_general_anti(E=energy, beta=0, delta_cp=radians(90))
prob_mue_matter_anti_NO2 = probability_general_anti(E=energy, beta=0, delta_cp=radians(-90))
prob_mue_matter_anti_IO = probability_general_anti(E=energy, beta=0, delta_cp=radians(90), MO='IO')
prob_mue_matter_anti_IO2 = probability_general_anti(E=energy, beta=0, delta_cp=radians(-90), MO='IO')
```

```python
plt.plot(energy, prob_mue_matter_NO, label='NO, matter')
plt.plot(energy, prob_mue_matter_IO, label='IO, matter')

plt.legend()
```

$ \textbf{Example #3}:$ General plot for different MO and $\delta_{CP}$

```python
rc('figure', figsize=(20, 9))
fig, [ax, ax2] = plt.subplots(1,2)
plt.axes(ax)
ax.fill_between(energy, prob_mue_matter_NO, prob_mue_matter_NO2, label=r'Normal ordering $\nu_\mu \rightarrow \nu_e$')
plt.plot(energy, prob_mue_matter_NO, label=r'NO, $\delta_{CP}=90$', color = 'blue', ls ='--')
plt.plot(energy, prob_mue_matter_NO2, label=r'NO, $\delta_{CP}=-90$', color = 'blue')

ax.fill_between(energy, prob_mue_matter_IO, prob_mue_matter_IO2, label=r'Inverse ordering $\nu_\mu \rightarrow \nu_e$')
ax.fill_between(energy, prob_mue_matter_IO2, prob_mue_matter_NO, label='Degeneracy region', color='saddlebrown')
plt.plot(energy, prob_mue_matter_IO, label=r'IO, $\delta_{CP}=90$', color = 'red', ls='--')
plt.plot(energy, prob_mue_matter_IO2, label=r'IO, $\delta_{CP}=-90$', color = 'red')
plt.xlabel('energy. GeV')
plt.ylabel(r'P($\nu_\mu \rightarrow \nu_e$)')
plt.legend()

plt.axes(ax2)

ax2.fill_between(energy, prob_mue_matter_anti_NO, prob_mue_matter_anti_NO2, label=r'Normal ordering $\bar{\nu}_\mu \rightarrow \bar{\nu}_e$')
plt.plot(energy, prob_mue_matter_anti_NO, label=r'NO, $\delta_{CP}=90$', color = 'blue', ls ='--')
plt.plot(energy, prob_mue_matter_anti_NO2, label=r'NO, $\delta_{CP}=-90$', color = 'blue')

ax2.fill_between(energy, prob_mue_matter_anti_IO, prob_mue_matter_anti_IO2, label=r'Inverse ordering $\bar{\nu}_\mu \rightarrow \bar{\nu}_e$')
plt.plot(energy, prob_mue_matter_anti_IO, label=r'IO, $\delta_{CP}=90$', color = 'red', ls='--')
plt.plot(energy, prob_mue_matter_anti_IO2, label=r'IO, $\delta_{CP}=-90$', color = 'red')

ax2.fill_between(energy, prob_mue_matter_anti_NO, prob_mue_matter_anti_IO2, label='Degeneracy region', color='saddlebrown')

plt.xlabel('energy, GeV')
plt.ylabel(r'P($\bar{\nu}_\mu \rightarrow \bar{\nu}_e$)')
plt.legend()
plt.savefig('ME')
```
